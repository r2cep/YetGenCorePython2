Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random

kanal = ["A Kanalı","B Kanalı","C Kanalı","D Kanalı"]
randomKanal = random.choice(kanal)
print(randomKanal)
kanalSayısı = len(kanal)
televizyonİnfosu = """
"Marka" : "Samsung"
Çıkış Yılı : 2023
Güç Tüketimi : 15V
Ekran Özelliği : Ultra Led 
"""
class Kumanda():
    def __init__(self,tvDurum = "kapalı",sesAyarları= "düşük ses"):
        self.tvDurum = tvDurum
        self.sesAyarları = sesAyarları
        self.kanalSayısıögrenme = kanalSayısı
        
    
    def tvAc(self):
        if (self.tvDurum == "kapalı"):
            print("TV açılıyor....")
            self.tvDurum = "açık"
        else:
            print("TV kapanıyor...")
            self.tvDurum = "kapalı"
            print("TV açık")
    def tvKapat(self):
        if (self.tvDurum == "kapalı"):
            print("TV ' yi açınız....")
        else:
            if (self.tvDurum == "kapalı"):
                print("TV zaten kapalı....")
            else:
                print("TV kapanıyor...")
                self.tvDurum = "kapalı"
    
    def sesAyarla(self):
        if self.tvDurum == "kapalı":
            print("Önce TV ' yi açınız.")
        else :
            mod = input("İstediğiniz ses seviyesini giriniz : (Varsıylan = düşük ses)")
            self.sesAyarları = "mod"
    
    def kanalEkle(self):
        if self.tvDurum == "kapalı":
            print("Önce TV 'yi açınız.")
        elif self.tvDurum == "açık":
            kanalEkleme = input("Eklemek istediğiniz kanalı giriniz : ")
            kanal.append(kanalEkleme)
    
    def kanalSayısı(self):
        if self.tvDurum == "kapalı":
            print("Önce TV ' yi açınız.")
        else:
            print(f"Kanal Sayısı : {kanalSayısı}")
    def rastgeleKanal(self):
        if self.tvDurum == "kapalı":
            print("Önce TV ' yi açınız.")
        else:
            print(f"Sizin için seçtiğimiz kanal : {randomKanal}")
    def televizyonİnfo(self):
        if self.tvDurum == "kapalı":
            print("Önce TV ' yi açınız.")
        else:
            print(televizyonİnfosu)


    

Kumanda = Kumanda()

print("""
Kahve Otomatı
1. TV Aç
2. TV Kapat
3. Ses Ayarları
4. Kanal Ekle   
5. Kanal Sayısı Öğrenme 
6. Rastgele Kanal Geçme
7. Televizyon Bilgileri
Çıkmak için 'q' ya basın.""")



while True:
    islem = input("işlemi seçiniz:")
    if (islem == "q"):
        print("Görüşmek üzere!")
        break
    elif (islem == "1"):
        Kumanda.tvAc()
    elif (islem == "2"):
        Kumanda.tvKapat()
    elif (islem == "3"):
        Kumanda.sesAyarla()
    elif (islem == "4"):
        Kumanda.kanalEkle()
    elif (islem == "5"):
        Kumanda.kanalSayısı()
    elif (islem == "6"):
        Kumanda.rastgeleKanal()
    elif(islem == "7"):
        Kumanda.televizyonİnfo()
    else:
        pass
```
