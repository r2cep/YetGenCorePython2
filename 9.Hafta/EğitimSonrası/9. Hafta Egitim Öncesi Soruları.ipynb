# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self,marka,yıl,beygir):
        self.marka = marka
        self.yıl = yıl
        self.beygir = beygir
class Bus(Vehicle):
    def __init__(self, marka, yıl, beygir,seating_capacity=50):
        super().__init__(marka, yıl, beygir)
        self.seating_capacity = seating_capacity
    def intro(self):
        print(f"The seating capacity of a bus is {self.seating_capacity}")
    
my_bus = Bus("Mercedes",2022,400)
my_bus.intro()
````

# Problem 2


School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python
class Vehicle:
    pass

class Bus(Vehicle):
    def seating_capacity(self, seating_capacity=50):
        print(f"The seating capacity of a bus is {seating_capacity} passengers")

class SchoolBus(Bus):
    pass

bus = SchoolBus()
print(isinstance(bus, Vehicle))
````
